```python
# Import

import pandas as pd
import numpy as np
```

```python
# Read measurements CSV file

csv_measurements = pd.read_csv('./Resources/hawaii_measurements.csv')
csv_measurements.head()
```

```python
# Clean Data above by removing NaN and 0

clean_csv_measurements = csv_measurements.dropna()
clean_csv_measurements = clean_csv_measurements[clean_csv_measurements.prcp != 0]
clean_csv_measurements.head()
```

```python
# Create new CSV file of clean measurements data

clean_csv_measurements.to_csv('clean_measurements.csv')
```

```python
# Read stations CSV file

csv_stations = pd.read_csv('./Resources/hawaii_stations.csv')
csv_stations.head()
```

```python
csv_stations.to_csv('clean_stations.csv')
```

```python
merged_data = pd.merge(clean_csv_measurements, csv_stations, on = 'station')
merged_data.head(10)
```
